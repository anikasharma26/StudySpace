```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
```

```python
y = df['logS']
```

```python
x = df.drop('logS',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_lr_train_pred = model.predict(x_train)
y_lr_test_pred = model.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
y_lr_test_pred
```

```python
# Sample input data from the question
sample_data = [[2.3765, 133.405, 0.0, 0.0]]

# Predict the LogS using the trained model
predicted_logS = model.predict(sample_data)
print("Predicted LogS:", predicted_logS[0])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=y_train,y=y_lr_train_pred,alpha=0.3)


plt.plot()
plt.ylabel('Predict logS')
plt.xlabel('Experimnetal logs')
```
